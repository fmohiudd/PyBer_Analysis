```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os

#Load in csv
ride_data_to_load = os.path.join("Resources", "PyBer_ride_data.csv")
pyber_ride_df = pd.read_csv(ride_data_to_load)
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y= "Avg. Fare ($USD)" )
plt.show()
```

```python
x_axis = np.arange(len(pyber_ride_df))
x_axis
```

```python
# Add x-ticks and tick locations. 
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]

# Plot the data. 
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
# Method #1: Plot a bar chart from the data frame. 

pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Mthod #2: Another way of plotting the bar graph. 
pyber_ride_df.plot(x = "Month", y = "Avg. Fare ($USD)", kind = "bar")
plt.show()
```

```python
# Skill drilll 
# Add error bars and caps
import statistics
# Calculate the standard deviation
y = pyber_ride_df["Avg. Fare ($USD)"]

stdev = statistics.stdev(y)
stdev
```
