```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# THIS IS THE MATLAB METHOD
# Set the x-axis to a list of strings for each month. 
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "July", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in the US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09 ]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# LINE CHART USING THE OBJECT-ORIENTED INTERFACE
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() <-- QUICK GRAPH USING OBJECT-OREINTED Method. 
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label = "Boston")
# Create the labels for the x and y axes. 
plt.xlabel("Date")
plt.ylabel("Fares($)")
#Set the y limit between 0 and 45. 
plt.ylim(0,45)
# Create a title.
plt.title("PyBer Far by Month")
# Add legend. 
plt.legend()
```

```python
# Create the plot. 
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth =2, label = 'Boston')
# Create labels for the x and y axes. 
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Set the y limit between 0 and 45. 
plt.ylim(0,45)

# SET X_LIM FROM jANRUARY THROUGH JUNE

#plt.xlim(0, 5)

# Create a title. 
plt.title("PyBer Fare by Month")
# Add a grid. 
plt.grid()
# Add the legend. 
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month. 
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "July", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in the US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09 ]

# LINE CHART USING THE OBJECT-ORIENTED INTERFACE
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, label = 'Boston')

ax.legend()
ax.grid()

#fig, ax = plt.subplots(figsize=(20, 10))
ax.set_ylim(0,45)

ax.set_xlabel('Months of the year')
ax. set_ylabel('Fares ($)')


# Method 2
#plt.subplots(figsize =(20,10))
```

```python
# Can also be written as (Object Oriented)
# Create the plot with ax.plt()
fig=plt.figure()
ax=fig.add_subplot()
ax.plot(x_axis, y_axis, marker = "D", c="g")
ax.set_xlabel("Months")
ax.set_ylabel("Fares($)")
ax.grid()
```
