```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# THIS IS THE MATLAB METHOD
# Set the x-axis to a list of strings for each month. 
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "July", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in the US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09 ]
```

```python
# Create a VERTICAL BAR
# Creat the plot
#plt.bar(x_axis, y_axis)
```

```python
# Create a VERTICAL BAR
# Creat the plot and add labels
plt.bar(x_axis, y_axis, color = "green", label = 'Boston')
#Create labels for  the x-axis and y-axis. 
plt.xlabel("Date")
plt.ylabel("Fares ($)")
# Create a title. 
plt.title("PyBer Fare by Month")
# Add legend.
plt.legend()
```

```python
# Create a HORIZONTAL BAR CHART
plt.barh(x_axis, y_axis)
```

```python
# Create a HORIZONTAL BAR CHART. INVERT THE y-axis LABEL so that DECEMBER IS AT THE TOP
plt.barh(x_axis, y_axis, color = "purple", label = 'Boston')
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.xlabel("Fares($)")
plt.ylabel("Months")

plt.legend()
```

```python
# THIS IS THE Object-Oriented approach
# Set the x-axis to a list of strings for each month. 
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "July", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in the US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09 ]
```

```python
# Create the phot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color = 'black',  label = 'New York')

ax.legend()
```

```python
# Create the plot with ax.plt()
fig, ax =plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
# Create the plot with ax.plt()
fig, ax =plt.subplots()
ax.barh(x_axis, y_axis, color = "purple", label = "Chicago")

ax.set_xlabel("Fare ($)")
ax.set_ylabel("Months")
ax.set_title("PyBer Fares by Month in Chicago")
ax.invert_yaxis()    # Invert y-axis labels


ax.legend()
```
